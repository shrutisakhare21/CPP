```python
# 6. Basic string compression
s = "aaabbcccaaa"
res, count = "", 1
for i in range(1, len(s)):
    if s[i] == s[i-1]:
        count += 1
    else:
        res += s[i-1] + str(count)
        count = 1
res += s[-1] + str(count)
print(res)
```

```python
# 20. Shuffle a list without built-in shuffle
import random
lst = [1,2,3,4,5]
for i in range(len(lst)-1,0,-1):
    j = random.randint(0, i)
    lst[i], lst[j] = lst[j], lst[i]
print(lst)
```

```python
# 31. Word frequency
words = ["apple", "banana", "apple", "orange", "banana"]
freq = {w: words.count(w) for w in set(words)}
print(freq)
```

```python
# 32. Merge two dictionaries (add common keys)
d1 = {'a':1,'b':2,'c':3}
d2 = {'b':3,'c':4,'d':5}
merged = {k: d1.get(k,0)+d2.get(k,0) for k in set(d1)|set(d2)}
print(merged)
```

```python
# 33. Access value in nested dictionary
def access_nested(d, keys):
    for key in keys:
        if isinstance(d, dict) and key in d:
            d = d[key]
        else:
            return None
    return d

nested = {'a': {'b': {'c': 42}}}
print(access_nested(nested, ['a','b','c']))
```

```python
# 34. Sort dictionary by values
d = {'a':3,'b':1,'c':2}
sorted_d = dict(sorted(d.items(), key=lambda x: x[1]))
print(sorted_d)
```

```python
# 35. Invert a dictionary
d = {'a':1,'b':2,'c':1}
inverted = {}
for k,v in d.items():
    inverted.setdefault(v, []).append(k)
print(inverted)
```
