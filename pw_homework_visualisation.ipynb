**BOKEH**

Q1 create bokeh plot displaying a sine wave.set x-values from 0-10 and y values as the sine of x

```python


from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Use Bokeh in notebook
output_notebook()

# Generate values
x = np.linspace(0, 10, 300)
y = np.sin(x)

# Create figure
p = figure(title="Sine Wave using Bokeh",
           x_axis_label="x",
           y_axis_label="sin(x)",
           width=600,
           height=400)

# Add sine curve
p.line(x, y, line_width=2)

# Display plot
show(p)
```

Q2 bokeh scatter plot using randomly generated x and y values.use different sizes and colors for the markers based on sizes and colors column

```python


from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
import pandas as pd
from bokeh.models import ColumnDataSource

# Enable output in notebook
output_notebook()

# Generate random data
np.random.seed(42)
x = np.random.rand(100) * 10
y = np.random.rand(100) * 10

sizes = np.random.randint(5, 25, 100)   # random marker sizes
colors = np.random.choice(["red", "blue", "green", "orange", "purple"], 100)  # random colors

# Dataset
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create figure
p = figure(title="Bokeh Scatter Plot with Sizes & Colors",
           x_axis_label="X Values",
           y_axis_label="Y Values",
           width=600,
           height=400)

# Add scatter plot
p.scatter("x", "y", size="sizes", color="colors", source=source, fill_alpha=0.6)

# Display
show(p)
```

Q3 Generate bokeh bar chart to represent count of fruits using following dataset

```python
from bokeh.plotting import figure, output_file, show

output_file("bar_chart.html")
fruits = ['Apple', 'Orange', 'Banana', 'Pears']
counts = [20, 25, 30, 35]

p = figure(x_range=fruits, title="Bar Chart", height=400)
p.vbar(x=fruits, top=counts, width=0.5, fill_color='pink')
show(p)
```

Q4 create a bokeh histogram to visualize distribution of given data

```python
import numpy as np
from bokeh.plotting import figure, output_file, show

output_file("histogram.html")

data_hist=np.random.randn(1000)
hist,edges=np.histogram(data_hist,bins=30)

p = figure(x_range=fruits, title="histogram")
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="skyblue", line_color="white")
show(p)
```

Q5 Generate a synthetic dataset with correlated features.Visualize the correlation matrix of dataset using heatmap

```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Create correlated features
np.random.seed(42)

# base feature
x1 = np.random.normal(0, 1, 500)
# correlated features
x2 = 0.8 * x1 + np.random.normal(0, 0.3, 500)
x3 = 0.6 * x1 + np.random.normal(0, 0.4, 500)
x4 = np.random.normal(0, 1, 500)  # independent feature

# Step 2: Form a DataFrame
df = pd.DataFrame({
    "Feature1": x1,
    "Feature2": x2,
    "Feature3": x3,
    "Feature4": x4
})

# Step 3: Compute correlation matrix
corr_matrix = df.corr()

# Step 4: Visualize using heatmap
plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix Heatmap")
plt.show()
```
